```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Create a dummy dataset

```python
df = pd.read_csv('../raw_data/Consumption_Cleaned.csv')
df
```

```python
df = df[df['EU?']]
df
```

```python
columns = ['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998',
         '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007',
         '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
         '2017', '2018', '2019', '2020']
df[columns]= df[columns].apply(pd.to_numeric, errors = 'coerce')
df
```

```python
df = df[df['energy_balance']=='Finalconsumption']
df
```

```python
data = pd.DataFrame(df[columns].sum(numeric_only=True, axis=0))

data
```

```python
data = data.rename(columns={0:'value'})
data
```

```python
data.columns
```

```python
df2 = data.copy();

for i in range(1, 11):
    df2[f't - {i}'] = df2['value'].shift(i)


df2.dropna(inplace=True)    
df2
```

```python
import pmdarima as pm
sarimax = pm.auto_arima(df2[['value']],
                           start_p=0, start_q=0,
                           test='adf',
                           max_p=5, max_q=5, m=1,
                           start_P=0, seasonal=False,
                           d=None, D=None, trace=True,
                           suppress_warnings=True, 
                           stepwise=True,
                           scoring='mse')
sarimax.aic()
```

```python
sarimax.plot_diagnostics(figsize=(10,10))
```

```python
sarimax.summary()
```

```python
sarimax.fit(df2[['value']])
```

```python
from prophet import Prophet
```

```python
df = data
df.reset_index(inplace=True)
df.rename(columns={'index':'ds', 'value': 'y'}, inplace=True)
df
```

```python
start = pd.to_datetime('2016, 1, 1')
end = start + pd.to_timedelta(52, 'W')
end
```

```python
df['ds'] = pd.to_datetime(df['ds'], format='%Y') + pd.to_timedelta(364, unit='D')
df
```

```python
train = df.iloc[:25]
test = df.iloc[25:]
train.shape, test.shape
```

```python
train
```

```python
model = Prophet(seasonality_mode='multiplicative',
                yearly_seasonality='auto',
                weekly_seasonality=False,
                daily_seasonality=False,
                interval_width=0.95)
model.fit(train)

# 135.92 --> yearly_seasonality=True
```

```python
horizon = 16
future = model.make_future_dataframe(periods=horizon, freq='Y')
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
test
```

```python
ax = model.plot(forecast);
plt.plot(test.ds, test.y, color = 'r')
```

```python
plt.figure(figsize=(15, 8))
plt.plot(train['ds'], train['y'], label="Train")
plt.plot(test['ds'], test['y'], label="Test")
plt.plot(forecast[-horizon:]['ds'], forecast[-horizon:]['yhat'], label="predictions")
plt.legend(loc="best")
plt.show()
```

```python
model.plot_components(forecast);
```

# CV

```python
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics
# Your code here
cv_results = cross_validation(model=model, initial=f'{365 * 20} days', horizon='365 days', period='180 days')
```

```python
cv_results
```

```python
df_p = performance_metrics(cv_results)
df_p
```
