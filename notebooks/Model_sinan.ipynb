```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Create a dummy dataset

```python
df = pd.read_csv('../raw_data/Consumption_Cleaned.csv')
df
```

```python
df = df[df['EU?']]
df
```

```python
columns = ['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998',
         '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007',
         '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016',
         '2017', '2018', '2019', '2020']
df[columns]= df[columns].apply(pd.to_numeric, errors = 'coerce')
df
```

```python
df = df[df['energy_balance']=='Finalconsumption']
df
```

```python
data = pd.DataFrame(df[columns].sum(numeric_only=True, axis=0))

data
```

```python
data = data.rename(columns={0:'value'})
data
```

```python
data.columns
```

```python
df2 = data.copy();

for i in range(1, 11):
    df2[f't - {i}'] = df2['value'].shift(i)


df2.dropna(inplace=True)    
df2
```

# Initialize the model

```python
plt.plot(df2.value)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df2.value, lags=10);
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(df2.value, lags=10, c='r');
```

```python
import pmdarima as pm
sarimax = pm.auto_arima(df2[['value']],
                           start_p=0, start_q=0,
                           test='adf',
                           max_p=5, max_q=5, m=1,
                           start_P=0, seasonal=False,
                           d=None, D=None, trace=True,
                           suppress_warnings=True, 
                           stepwise=True)
```

```python
sarimax.plot_diagnostics(figsize=(10,10))
```

```python
sarimax.summary()
```

```python
sarimax.fit()
```
