## Consumption Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Name • Type (categorical, int/float, bounded/unbounded, text, structured, etc.)

```python
consumption_df = pd.read_csv('../raw_data/ElecConsumption.csv')
consumption_df
```

```python
consumption_df.info()
```

--> data from 1990 to 2013 and from 2019 to 2020: **object**, but data from 2014 to 2018: **float**

```python
consumption_df.describe()
```

```python
sum(consumption_df.duplicated())
```

--> no duplicates

### % of missing values

```python
consumption_df.isna().sum().sum()
```

--> no na values

```python
consumption_df.apply(lambda serie: serie.eq(': ').sum(), axis=0)
```

```python
consumption_df.groupby('Alpha_2_code').count()
```

--> 41 countries, EU27, EA19

```python
consumption_df_missing_val = consumption_df.groupby('Alpha_2_code').aggregate(lambda serie: serie.eq(': ').sum())
consumption_df_missing_val
```

```python
missing_data_countries = consumption_df_missing_val[consumption_df_missing_val.sum(axis=1) != 0]
missing_data_countries
```

```python
consumption_df_float = consumption_df.loc[:, '1990':'2020'].replace(': ',np.nan).astype(float)
consumption_df_float
```

```python
consumption_df_float_complete = pd.merge(consumption_df.loc[:, :'Alpha_2_code'], consumption_df_float, left_index=True, right_index=True)
consumption_df_float_complete
```

```python
consumption_df_float_complete.info()
```

```python
consumption_df_float_complete.describe()
```

### Noisiness and type of noise (stochastic, outliers, rounding errors, etc.)

```python
consumption_df_country_sums = consumption_df_float_complete.groupby('Alpha_2_code').sum()
consumption_df_country_sums
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(consumption_df_country_sums['2015'])
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(consumption_df_country_sums['2004'])
```

```python
consumption_df_country_sums.T.plot(figsize=(20,15))
```

```python
consumption_df_sums_only_countries = consumption_df_country_sums.reset_index()[~consumption_df_country_sums.reset_index()['Alpha_2_code'].isin(['EA19', 'EU27_2020'])]
consumption_df_sums_only_countries
```

```python
mask = consumption_df_sums_only_countries.loc[:, '1990':'2020'] == 0
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(consumption_df_sums_only_countries.loc[:, '1990':'2020'], yticklabels=consumption_df_sums_only_countries['Alpha_2_code'], mask=mask, cmap="viridis")
```

```python
consumption_df_sums_only_countries.set_index('Alpha_2_code').T.plot(figsize=(20,15))
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(consumption_df_sums_only_countries.set_index('Alpha_2_code')['2015'], bins=20)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.histplot(consumption_df_sums_only_countries.set_index('Alpha_2_code')['2004'], bins=20)
```
