### Name • Type (categorical, int/float, bounded/unbounded, text, structured, etc.)

```python
consumption_df = pd.read_csv('../raw_data/ElecConsumption.csv')
consumption_df
```

```python
consumption_df.isna().sum().sum()
```

```python
consumption_df.apply(lambda serie: serie.eq(': ').sum(), axis=0)
```

#### Group by country

```python
consumption_df.groupby('Alpha_2_code').count()
```

--> 41 countries, EU27, EA19

```python
consumption_df_missing_val = consumption_df.groupby('Alpha_2_code').aggregate(lambda serie: serie.eq(': ').sum())
consumption_df_missing_val
```

```python
missing_data_countries = consumption_df_missing_val[consumption_df_missing_val.sum(axis=1) != 0]
missing_data_countries
```

```python
consumption_df_float = consumption_df.loc[:, '1990':'2020'].replace(': ',np.nan).astype(float)
consumption_df_float
```

```python
consumption_df_float_complete = pd.merge(consumption_df.loc[:, :'Alpha_2_code'], consumption_df_float, left_index=True, right_index=True)
consumption_df_float_complete
```

```python
consumption_df_float_complete.info()
```

```python
consumption_df_float_complete.describe()
```

```python
consumption_df_float_complete_only_countries = consumption_df_float_complete.reset_index()[~consumption_df_float_complete.reset_index()['Alpha_2_code'].isin(['EA19', 'EU27_2020'])]
consumption_df_float_complete_only_countries
```
